```python
with open("semlink-2.txt", "r") as f:
    semlink_data = [line.split() for line in f.readlines()]

fn = []    
non_fn = []

for line in semlink_data:
    if line[5] == "None":
        non_fn.append(line)
    else:
        fn.append(line)
```

```python
fn_x = []
fn_error = []
for line in fn:
    frame = line[5]
    links = []
    for feature in line[5:]:
        if "=" in feature:
            mapping = feature.split("=")
            pb_role = mapping[0].split("-")[-1]
            if ";" in mapping[1]:
                fn_role = mapping[1].split(";")[-1]
            else:
                fn_role = mapping[1]
            link = (pb_role, fn_role)
            links.append(link)
    if links != []:
        fn_x.append((frame, links))
    else:
        fn_error.append(line)
            
    
                
```

```python
pb2fn = {}
fn2pb = {}
for x in fn_x:
    for pb, fn in x[1]:
        if pb in pb2fn:
            if fn not in pb2fn[pb]:
                pb2fn[pb].append(fn)
        else:
            lst=[fn]
            pb2fn[pb]=lst
        if fn in fn2pb:
            if pb not in fn2pb[fn]:
                fn2pb[fn].append(pb)
        else:
            lst=[pb]
            fn2pb[fn]=lst
```
