```python
import operator
```

```python
with open("../data/train.csv") as f:
    train_data = [[field.split(" ") for field in line.split("\t")] for line in f.readlines()]
```

```python
with open("../data/test.csv") as f:
    test_data = [[field.split(" ") for field in line.split("\t")] for line in f.readlines()]
```

```python
labels_in_train = {}

for line in train_data:
    for label in line[2]:
        label = label.replace("\n", "")
        if label in labels_in_train:
            labels_in_train[label] +=1 
        else:
            labels_in_train[label] = 1
```

```python
labels_in_test = {}

for line in test_data:
    for label in line[2]:
        label = label.replace("\n", "")
        if label in labels_in_test:
            labels_in_test[label] +=1 
        else:
            labels_in_test[label] = 1
```

```python
total = sum(labels_in_train.values())
train_as_list = list(labels_in_train.items())
train_as_list.sort(key = operator.itemgetter(1), reverse = True)
print("TRAIN")
for x in train_as_list:
    print(f"{x[0]}\t{x[1]}\t{round((x[1]/total)*100, 2)} %")
```

```python
total = sum(labels_in_test.values())
test_as_list = list(labels_in_test.items())
test_as_list.sort(key = operator.itemgetter(1), reverse = True)
print("TEST")
for x in test_as_list:
    print(f"{x[0]}\t{x[1]}\t{round((x[1]/total)*100, 2)} %")
```
